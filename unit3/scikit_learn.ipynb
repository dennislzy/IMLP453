```python


import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# 簡化的數據集
data = {
    '城市': ['台北', '高雄', '台中', '台北', '高雄'],
    '教育程度': ['高中', '大學', '碩士', '大學', '高中'],
    '年齡': [25, 30, 35, 28, 33],
    '購買': [0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
print("原始數據:")
print(df)

# 2. Label Encoding (有序類別)

print("\n🏷️ Label Encoding - 教育程度 (有序)")
print("-" * 30)

# 創建編碼器
education_encoder = LabelEncoder()

# 使用 fit_transform
df['教育程度_encoded'] = education_encoder.fit_transform(df['教育程度'])

print("\n結果:")
print(df[['教育程度', '教育程度_encoded']])

# 3. One-Hot Encoding (無序類別)

print("\n🔥 One-Hot Encoding - 城市 (無序)")
print("-" * 30)

# 創建編碼器
city_encoder = OneHotEncoder(sparse_output=False, drop='first')

# 使用 fit_transform
city_encoded = city_encoder.fit_transform(df[['城市']])

# 獲取特徵名稱
feature_names = city_encoder.get_feature_names_out(['城市'])
print(f"生成特徵: {list(feature_names)}")

# 創建 DataFrame
city_df = pd.DataFrame(city_encoded, columns=feature_names)

print("\n結果:")
result = pd.concat([df[['城市']], city_df], axis=1)
print(result)

# 4. 最終合併結果

print("\n✅ 最終編碼結果")
print("-" * 20)

# 合併所有特徵
df_final = pd.concat([
    df[['年齡']],                           # 原始數值特徵
    df[['教育程度_encoded']],               # Label encoded
    city_df,                                # One-hot encoded
    df[['購買']]                           # 目標變數
], axis=1)

print(df_final)

# 5. 新數據處理

print("\n🆕 處理新數據")
print("-" * 15)

# 新數據
new_data = pd.DataFrame({
    '城市': ['台中'],
    '教育程度': ['大學'],
    '年齡': [29]
})

print("新數據:")
print(new_data)

# 使用已訓練的編碼器 (注意：只用 transform，不用 fit_transform)
new_education = education_encoder.transform(new_data['教育程度'])
new_city = city_encoder.transform(new_data[['城市']])

print(f"\n編碼結果:")
print(f"教育程度 '大學' → {new_education[0]}")
print(f"城市 '台中' → {new_city[0]}")

# 6. 核心要點

print("\n💡 核心要點")
print("="*20)
print("Label Encoding:")
print("  • 用於有序類別 (高中 < 大學 < 碩士)")
print("  • 結果是數字 (0, 1, 2...)")

print("\nOne-Hot Encoding:")
print("  • 用於無序類別 (城市名稱)")
print("  • 結果是向量 ([1,0], [0,1]...)")

print("\nfit_transform vs transform:")
print("  • 訓練時: fit_transform()")
print("  • 新數據: transform()")
```
