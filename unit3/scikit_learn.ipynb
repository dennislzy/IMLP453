```python


import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# ç°¡åŒ–çš„æ•¸æ“šé›†
data = {
    'åŸå¸‚': ['å°åŒ—', 'é«˜é›„', 'å°ä¸­', 'å°åŒ—', 'é«˜é›„'],
    'æ•™è‚²ç¨‹åº¦': ['é«˜ä¸­', 'å¤§å­¸', 'ç¢©å£«', 'å¤§å­¸', 'é«˜ä¸­'],
    'å¹´é½¡': [25, 30, 35, 28, 33],
    'è³¼è²·': [0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)
print("åŸå§‹æ•¸æ“š:")
print(df)

# 2. Label Encoding (æœ‰åºé¡åˆ¥)

print("\nğŸ·ï¸ Label Encoding - æ•™è‚²ç¨‹åº¦ (æœ‰åº)")
print("-" * 30)

# å‰µå»ºç·¨ç¢¼å™¨
education_encoder = LabelEncoder()

# ä½¿ç”¨ fit_transform
df['æ•™è‚²ç¨‹åº¦_encoded'] = education_encoder.fit_transform(df['æ•™è‚²ç¨‹åº¦'])

print("\nçµæœ:")
print(df[['æ•™è‚²ç¨‹åº¦', 'æ•™è‚²ç¨‹åº¦_encoded']])

# 3. One-Hot Encoding (ç„¡åºé¡åˆ¥)

print("\nğŸ”¥ One-Hot Encoding - åŸå¸‚ (ç„¡åº)")
print("-" * 30)

# å‰µå»ºç·¨ç¢¼å™¨
city_encoder = OneHotEncoder(sparse_output=False, drop='first')

# ä½¿ç”¨ fit_transform
city_encoded = city_encoder.fit_transform(df[['åŸå¸‚']])

# ç²å–ç‰¹å¾µåç¨±
feature_names = city_encoder.get_feature_names_out(['åŸå¸‚'])
print(f"ç”Ÿæˆç‰¹å¾µ: {list(feature_names)}")

# å‰µå»º DataFrame
city_df = pd.DataFrame(city_encoded, columns=feature_names)

print("\nçµæœ:")
result = pd.concat([df[['åŸå¸‚']], city_df], axis=1)
print(result)

# 4. æœ€çµ‚åˆä½µçµæœ

print("\nâœ… æœ€çµ‚ç·¨ç¢¼çµæœ")
print("-" * 20)

# åˆä½µæ‰€æœ‰ç‰¹å¾µ
df_final = pd.concat([
    df[['å¹´é½¡']],                           # åŸå§‹æ•¸å€¼ç‰¹å¾µ
    df[['æ•™è‚²ç¨‹åº¦_encoded']],               # Label encoded
    city_df,                                # One-hot encoded
    df[['è³¼è²·']]                           # ç›®æ¨™è®Šæ•¸
], axis=1)

print(df_final)

# 5. æ–°æ•¸æ“šè™•ç†

print("\nğŸ†• è™•ç†æ–°æ•¸æ“š")
print("-" * 15)

# æ–°æ•¸æ“š
new_data = pd.DataFrame({
    'åŸå¸‚': ['å°ä¸­'],
    'æ•™è‚²ç¨‹åº¦': ['å¤§å­¸'],
    'å¹´é½¡': [29]
})

print("æ–°æ•¸æ“š:")
print(new_data)

# ä½¿ç”¨å·²è¨“ç·´çš„ç·¨ç¢¼å™¨ (æ³¨æ„ï¼šåªç”¨ transformï¼Œä¸ç”¨ fit_transform)
new_education = education_encoder.transform(new_data['æ•™è‚²ç¨‹åº¦'])
new_city = city_encoder.transform(new_data[['åŸå¸‚']])

print(f"\nç·¨ç¢¼çµæœ:")
print(f"æ•™è‚²ç¨‹åº¦ 'å¤§å­¸' â†’ {new_education[0]}")
print(f"åŸå¸‚ 'å°ä¸­' â†’ {new_city[0]}")

# 6. æ ¸å¿ƒè¦é»

print("\nğŸ’¡ æ ¸å¿ƒè¦é»")
print("="*20)
print("Label Encoding:")
print("  â€¢ ç”¨æ–¼æœ‰åºé¡åˆ¥ (é«˜ä¸­ < å¤§å­¸ < ç¢©å£«)")
print("  â€¢ çµæœæ˜¯æ•¸å­— (0, 1, 2...)")

print("\nOne-Hot Encoding:")
print("  â€¢ ç”¨æ–¼ç„¡åºé¡åˆ¥ (åŸå¸‚åç¨±)")
print("  â€¢ çµæœæ˜¯å‘é‡ ([1,0], [0,1]...)")

print("\nfit_transform vs transform:")
print("  â€¢ è¨“ç·´æ™‚: fit_transform()")
print("  â€¢ æ–°æ•¸æ“š: transform()")
```
