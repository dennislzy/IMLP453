```python
import numpy as np
import matplotlib.pylab as pyd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x = np.linspace(0,5,50)
y = 1.2*x+0.8+0.6*np.random.randn(50)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=20)
pyd.scatter(x_train,y_train)
```

```python

regre = LinearRegression()
x_2d = x_train.reshape(-1,1)
regre.fit(x_2d,y_train)
y_predit = regre.predict(x_2d)
print('相關係數',regre.coef_)
print('截距',regre.intercept_)
```

```python
#計算MSE
MSE = np.sum((y_predit-y_train)**2)/len(y_train)
MSE
```

```python
pyd.scatter(x_train,y_train)
pyd.plot(x_train,y_predit,"g")
```

```python
#test datas
x_test_2d = x_test.reshape(-1,1)
y_test_predit = regre.predict(x_test_2d)
MSE_TEST = np.sum((y_test_predit-y_test)**2)/len(y_test)
MSE_TEST
```

```python
#模型預測train data和test data 做比較
title='f(x)=1.2x+0.8'+',mse:'+str(round(MSE_TEST,2))
pyd.title(title)
pyd.xlabel("x")
pyd.ylabel("y")
pyd.scatter(x_test,y_test,alpha=0.8)
pyd.scatter(x_test,y_test_predit,c="y",alpha=0.8) 
```
