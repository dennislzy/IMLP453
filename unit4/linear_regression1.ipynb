```python
import numpy as np
import matplotlib.pylab as pyd
import pandas as pd
x = np.linspace(0,5,50)
y =1.2*x+2+1*np.random.randn(50)
pyd.scatter(x,y)
pyd.plot(x,1.2*x+2,"y")
```

```python
from sklearn.linear_model import LinearRegression
x_2 = x.reshape(-1,1)
regre = LinearRegression()
regre.fit(x_2,y)
y_2 = regre.predict(x_2)
pyd.scatter(x,y) #原本的數據
pyd.plot(x,y_2)

#計算mse值
MSE = (np.sum((y_2-y)*(y_2-y)))/len(y)
MSE
```

```python
x_info = np.linspace(0,5,50)
y_info = 2*x_info+0.7*np.random.randn(50) #加上noice來模擬誤差
x_info_2d = x_info.reshape(-1,1)
model = regre.fit(x_info_2d,y_info)
y_info_predit = model.predict(x_info_2d)
pyd.scatter(x_info,y_info)
pyd.plot(x_info,y_info_predit)
```
